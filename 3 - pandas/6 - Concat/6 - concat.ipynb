```python
# import library

import pandas as pd 
```

```python
# create dataframe 1 

indian_weather = {
    "City":["Mumbai","Delhi","Banglore"],
    "Temp":[32 , 45 , 30],
    "Humidity":[80 , 60 , 78]
}

# convert dictionary to dataframe 
df = pd.DataFrame(indian_weather)
df
```

```python
# create dataframe 2

us_weather = {
    "City":["New Yourk","chicago","Orlando"],
    "Temp":[21 , 12 , 24],
    "Humidity":[68 , 60 , 75]
}

# convert dictionary to dataframe 
df1 = pd.DataFrame(us_weather)
df1
```

```python
df
```

```python
df1
```

```python
# concat two dataframes 
# concat by columns 

df2 = pd.concat([df,df1],axis = 1)
df2
```

```python
# concat two dataframes 
# concat by rows 

df3 = pd.concat([df,df1],axis = 0)
df3
```

```python
# concat two dataframes 
# concat by rows 

df4 = pd.concat([df,df1],axis = 0 , ignore_index = True)
df4
```

```python
# Concatenation and keys 

df5 = pd.concat([df,df1],axis = 0 ,keys = ["India","Us"])
df5
```
