```python
# import library

import pandas as pd 
import numpy as np
```

```python
# import data 

df = pd.read_csv("weather_data1.csv")
df
```

```python
df.info()
```

```python
# check missing values

df.isnull().sum()
```

```python
# handle data using replace method 

# replace data

df1 = df.replace(-99999 , np.NaN) # takes 2 arguments ( current data ,want to replace )
df1
```

```python
df.isnull().sum()
```

```python
df2 = df1.fillna(0)
df2
```

```python
# Replace multiple values 

df
```

```python
df3 = df.replace([-99999 , -88888] , np.NaN)
df3
```

```python
# replace data by columns 

df4 = df.replace({
    "temperature": -99999,
    "windspeed" : [-99999 ,-88888 ],
    "event" : "0"
} , np.NaN)
df4
```

```python
# Replace by mapping 

# dataframe
df
```

```python
df5 = df.replace({
    -99999 : np.NaN,
    -88888 : 1,
    "0":"No Events"
})
df5
```
