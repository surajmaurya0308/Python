```python
# fill mean values 

df
```

```python
# temparary fill NaN to 0

t_df = df.fillna(0)
t_df
```

```python
# calculate mean for temp or windspeed 

temp_col = t_df.temperature.values
temp_col
```

```python
teamp_mean = temp_col.mean()
teamp_mean = round(teamp_mean)
teamp_mean
```

```python
# mean of windspeed 

ws_col = t_df.windspeed.values
mean_ws = ws_col.mean()
mean_ws = round(mean_ws)
mean_ws
```

```python
df3 = df.fillna({
    "temperature" :teamp_mean,
    "windspeed" : mean_ws
})
df3
```
