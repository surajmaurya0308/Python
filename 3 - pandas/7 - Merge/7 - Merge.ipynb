```python
# import library

import pandas as pd 
```

```python
# create dataframe 1

df1 = pd.DataFrame({
    "City":["New Yourk","Chicago","orlando"],
    "Temp":[21,14,25]
})
df1
```

```python
# create dataframe 2

df2 = pd.DataFrame({
    "City":["New Yourk","Chicago","orlando"],
    "humidity":[65,68,75]
})
df2
```

```python
# concat 

concat_data = pd.concat([df1,df2],axis = 0,ignore_index = True)
concat_data
```

```python
#  sir mera city name alag hua toh

# create dataframe 1

df_1 = pd.DataFrame({
    "City":["New Yourk","Chicago","Orlando" ,"Washington DC"],
    "Temp":[21,14,25 , 20]
})
df_1
```

```python
# create dataframe 2

df_2 = pd.DataFrame({
    "City":["Chicago","New Yourk",'San diago' ],
    "Humidity":[65 , 68,75]
})
df_2
```

```python
df_1
```

```python
df_2
```

```python
# merge  data 

merge_df = pd.merge(df_1,df_2 , on = "City")
merge_df
```

```python
# Joins 

#inner join
inner_join = pd.merge(df_1 , df_2 , on = "City" , how = "inner")
inner_join
```

```python
#outer join
outer_join = pd.merge(df_1 , df_2 , on = "City" , how = "outer")
outer_join
```

```python
#left join
left_join = pd.merge(df_1 , df_2 , on = "City" , how = "left")
left_join
```

```python
#right join
right_join = pd.merge(df_1 , df_2 , on = "City" , how = "right")
right_join
```

```python
# indicator Flage 

#outer join
outer_join = pd.merge(df_1 , df_2 , on = "City" , how = "outer" , indicator = True)
outer_join
```
