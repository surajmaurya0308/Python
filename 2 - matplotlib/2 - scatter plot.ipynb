```python
# # create data 

day = ['Monday','Tuesday','Wednesday','Thuresday','Friday','Saturday','Sunday']

min_temp = [23 , 27 , 24 , 21 , 20 , 25 , 24]
avg_temp = [26 , 31 , 29 , 26 , 27 , 30 ,28]
max_temp = [29 , 34 , 32 , 31 , 32 , 38 , 33]
```

```python
# plot data 

plt.title('Weather Data ')
plt.xlabel('Day')
plt.ylabel('Temprature')

plt.scatter(day , min_temp ,marker = "*" ,color = 'blue', label = "min_temp")
plt.scatter(day , avg_temp ,marker = "*" ,color = 'green' ,label = "avg_temp")
plt.scatter(day , max_temp ,marker = "*" ,color = 'red', label = "max_temp")

plt.legend(shadow = True)
```
