```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
X = [1,2,3,4,5,6,7]
y = [50,51,52,48,47,49,46]

```

```python
plt.title("weather data ")

plt.xlabel("Day")
plt.ylabel("Temprature")

#supported values are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'

plt.plot(X , y , color = "red",linewidth = 1,linestyle = 'dotted')
```

```python
plt.title("weather data ")

plt.xlabel("Day")
plt.ylabel("Temprature")

#supported values are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'

plt.plot(X , y , color = "red",marker = "*",markersize = 10 , markerfacecolor = 'blue')
```

# plot multile graph

```python
# import library

import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
# create data 

day = ['Monday','Tuesday','Wednesday','Thuresday','Friday','Saturday','Sunday']

min_temp = [23 , 27 , 24 , 21 , 20 , 25 , 24]
avg_temp = [26 , 31 , 29 , 26 , 27 , 30 ,28]
max_temp = [29 , 34 , 32 , 31 , 32 , 38 , 33]
```

```python
# plot data 

plt.title('Weather Data ')
plt.xlabel('Day')
plt.ylabel('Temprature')

plt.plot(day , min_temp ,marker = "*" , label = "min_temp")
```

```python

plt.title('Weather Data ')
plt.xlabel('Day')
plt.ylabel('Temprature')

plt.plot(day , avg_temp ,marker = "*" , label = "avg_tem")
```

```python
plt.title('Weather Data ')
plt.xlabel('Day')
plt.ylabel('Temprature')

plt.plot(day , max_temp ,marker = "*" , label = "max_tem")
```

```python
# all in one graph 

# plot data 

plt.title('Weather Data ')
plt.xlabel('Day')
plt.ylabel('Temprature')

plt.plot(day , min_temp ,marker = "*" ,color = 'blue', label = "min_temp")
plt.plot(day , avg_temp ,marker = "*" ,color = 'green' ,label = "avg_temp")
plt.plot(day , max_temp ,marker = "*" ,color = 'red', label = "max_temp")

plt.legend(shadow = True)
```
