```python
# import library 

import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
# data 

month_exp_labels = ["Home Rent" , 'Food','Phone/Internet Bill','bike','Other Utilities']
exp_value = [7000 ,6000 , 4000 , 3500 , 2000]
```

```python
# plot pie chart 

plt.pie(exp_value , labels = month_exp_labels)
plt.show()
```

```python
# plot pie chart 

plt.pie(exp_value , labels = month_exp_labels , autopct = "%.0f%%")
plt.show()
```

```python
# plot pie chart 

plt.pie(exp_value , labels = month_exp_labels , autopct = "%.0f%%", radius = 1.5,shadow = True)
plt.show()
```

```python
# plot pie chart 

plt.pie(exp_value , labels = month_exp_labels , autopct = "%.0f%%",explode = [0,0,0.2,0,0.3])
plt.show()
```

```python
# plot pie chart

colors = ['red','green','purple','white','orange']

plt.pie(exp_value , labels = month_exp_labels , autopct = "%.0f%%",colors =colors )
plt.show()
```
