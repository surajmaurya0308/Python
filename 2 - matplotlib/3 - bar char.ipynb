```python
# import library 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# simple barchat showing revenue of major tech companies 

company = ['Google','Amazon','microsoft','Facebook']
revenue = [90 , 136 , 100,60]
```

```python
# plot bar chart

plt.title("Us Tech Company Stock")
plt.xlabel("Company")
plt.ylabel('revenue')

plt.bar(company , revenue)
```

```python
# add profit of the company 

profit = [40 , 45 ,34 , 26]
```

```python
# plot bar chart

plt.title("Us Tech Company Stock")
plt.xlabel("Company")
plt.ylabel('revenue')

plt.bar(company , revenue , label = "Revenue")
plt.bar(company , profit , label = "Profit")
plt.legend()
```

```python
# create array with range of company 
import numpy as np

X_pos = np.arange(len(company))
X_pos
```

```python
# plot bar chart

plt.title("Us Tech Company Stock")
plt.xlabel("Company")
plt.ylabel('revenue')

plt.xticks(X_pos , company )

plt.bar(X_pos - 0.2 , revenue,width = 0.4 , label = "Revenue")
plt.bar(X_pos + 0.2 , profit ,width = 0.4 , label = "Profit")
plt.legend()
```
